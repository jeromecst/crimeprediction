```python
import pandas as pd
pd.set_option('display.width', None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use(['seaborn-bright','classic'])
```

```python
data = pd.read_csv('Crimes1M.csv')
data.head(15)
```

```python
data.columns
```

```python
data.isnull().sum() 
```

```python
data.describe(include = 'all')
```

```python
data['Arrest'] = np.where(data['Arrest']==False,1,0)
data.head(15)
```

```python
features = list(data.columns[1:7])
```

```python
for i in features :
    plt.figure(figsize=(8,3))
    plt.style.use(['seaborn-bright','dark_background'])
    sns.countplot(x = data[i],data = data, hue = 'Arrest', palette='hsv')
    plt.title(i,fontsize=20,color='w')
```
